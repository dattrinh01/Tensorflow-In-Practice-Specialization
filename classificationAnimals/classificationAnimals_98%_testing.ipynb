```
model = tf.keras.models.load_model("classificationAnimals.h5")
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
className = ["dog", "horse","elephant", "butterfly",  "chicken",  "cat", "cow",  "sheep", "squirrel", "spider"]
```

```
uploaded = files.upload()
for k in uploaded.keys():
    path = "./" + k
    img = image.load_img(path, color_mode = "rgb", target_size=(100,100))
    x = image.img_to_array(img) / 255.0
    x = np.expand_dims(x, axis = 0)
    images = np.vstack([x])
    result = model.predict(images)
    predict_labels = np.argmax(result)
    plt.imshow(img)
    print("This is a ",className[predict_labels])
```
